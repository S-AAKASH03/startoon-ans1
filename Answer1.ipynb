```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from the files
data1 = np.loadtxt('/content/Data_1.txt')
data2 = np.loadtxt('/content/Data_2.txt')

# Function to find maxima and minima
def find_peaks(data):
    maxima = []
    minima = []
    for i in range(1, len(data) - 1):
        if data[i] > data[i - 1] and data[i] > data[i + 1]:
            maxima.append(i)
        if data[i] < data[i - 1] and data[i] < data[i + 1]:
            minima.append(i)
    return maxima, minima

# Find maxima and minima for both datasets
maxima1, minima1 = find_peaks(data1)
maxima2, minima2 = find_peaks(data2)

# Plotting the results
def plot_data(data, maxima, minima, title):
    plt.plot(data, label='Signal')
    plt.scatter(maxima, data[maxima], color='red', label='Maxima')
    plt.scatter(minima, data[minima], color='blue', label='Minima')
    plt.xlabel('Index')
    plt.ylabel('Signal Value')
    plt.title(title)
    plt.legend()
    plt.show()

plot_data(data1, maxima1, minima1, 'Data Set 1')
plot_data(data2, maxima2, minima2, 'Data Set 2')

# Printing the indices of the peaks
print("Maxima indices for Data Set 1:", maxima1)
print("Minima indices for Data Set 1:", minima1)
print("Maxima indices for Data Set 2:", maxima2)
print("Minima indices for Data Set 2:", minima2)
```
